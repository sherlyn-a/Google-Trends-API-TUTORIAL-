```python
from pytrends.request import TrendReq 
%matplotlib inline
```

```python
pytrends = TrendReq(hl='en-GB',
                    tz=360,
                    timeout=(10,25),
                    retries=2,
                    backoff_factor=0.1)
```

```python
kw_list = ['life insurance', 'health insurance', 'auto insurance', 'home insurance']
```

```python
pytrends.build_payload(kw_list,
                       cat=38,
                       timeframe='today 5-y',
                       geo='CA-ON',
                       gprop='')
                      
```

```python
trends_data = pytrends.interest_over_time()
```

```python
trends_data.plot(figsize = (15,5))
```
